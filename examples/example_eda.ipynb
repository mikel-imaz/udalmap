# Exploratory Data Analysis example

```python
# Import module
from udalmap.utils import UdmDf

# Instantiate class
udm = UdmDf()
```

## Explore indicators
Call `.find()` method to find all indicators.

```python
# Get all indicators into a dataframe
df = udm.find()
df
```

Once in a Pandas dataframe, we can easily manipulate the dataset.

For example, we can count the number of indicators by subgroup and see the overall structure:

```python
# Compute number of indicators per group and subgroup
df.groupby(["group", "subgroup"])["indicator"].count()
```

Or print id and descriptions of indicators in a certain group or subgroup:

```python
# Show indicators for a defined subgroup
subgroup = "E.3: Tejido empresarial"
print(df.loc[df["subgroup"] == subgroup, "indicator"].values)
```

## Get indicator data
Once we have identified our indicator of interest, we can call the `.get()` method with indicator `id` followed by one of these arguments:
- `entities`: provinces of the Basque Autonomous Community, together with the whole community itself (CAE)
- `regions`: areas throughout the Basque Autonomous Community
- `municipalities`: towns in the Community

```python
# Get data for indicator and "entities"
udm.get("17", "entities")
```

```python
# Get data for indicator and "regions"
udm.get("17", "regions")
```

```python
# Get data for indicator and "municipalities"
udm.get("17", "municipalities")
```

## Plot indicator data
You can conveniently plot the data with the `.plot()` method.

```python
udm.plot("17", "entities")
```

If regions and municipalities are to be plotted, then the graph gets overwhelmed.

```python
# Indicator plotted across regions without filter
udm.plot("17", "regions")
```

Use parameter `filters` with a list of the elements you wish to plot.

```python
# Select regions
print(udm.get("17", "regions").index.values)
```

```python
# Indicator plotted across regions with filter
udm.plot("17", "regions", filters=["Goierri",
                                   "Gernika-Bermeo",
                                   "Donostialdea / Donostia-San Sebastián",
                                   "Bilbo Handia / Gran Bilbao",
                                  ])
```

The same in the case of municipalities:

```python
# Select municipalities
print(udm.get("17", "municipalities").index.values)
```

```python
# Indicator plotted across municipalities with filter
udm.plot("17", "municipalities", filters=["Urretxu",
                                          "Zumarraga",
                                          "Donostia / San Sebastián",
                                          "Bilbao",
                                  ])
```

These out-of-the-box plots can provide insights into null values, erroneous data, and other issues, helping to guide data preprocessing before conducting a full-fledged analysis.
